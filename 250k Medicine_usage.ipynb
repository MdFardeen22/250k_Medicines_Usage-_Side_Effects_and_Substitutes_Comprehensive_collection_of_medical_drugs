```python
import pandas as pd
```

```python
fname = "250k Medicines Usage, Side Effects and Substitutes.csv"
raw_data = pd.read_csv(fname)
df= raw_data.copy()
df.head()
```

```python
df.info()
```

```python
df = df.drop("id",axis =1)
df.head()
```

```python
df.duplicated().any()
```

```python
# Drop duplicated rows, keeping only the first occurrence
df = df.drop_duplicates()
#checking again is there any duplicated rows after deletetion
df.duplicated().any()
```

```python
df.isnull().sum().head()
```

```python
#making a dataframe with the missing values of the dataset
null_counts = df.isnull().sum()
missing_value = pd.DataFrame(null_counts, columns=['null_count']).reset_index()
missing_value = missing_value.rename(columns={'index': 'Column'})
# Drop the first two row namem,as name has no missing value
missing_value = missing_value.drop([0]).reset_index(drop=True)
missing_value.head()
```

```python
#creating a barplot with the missing value
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14, 6))
sns.barplot(x='Column', y='null_count', data=missing_value, color ='red')
plt.xlabel('columns')
plt.ylabel('missing_value')
plt.title("missing value graph")
plt.xticks(rotation=90) # Rotate x-axis labels if necessary
plt.show()
# Show the plot
plt.show()
```

```python
#filling the missing values with unknown
df.fillna("UnKnown",inplace = True)
#checking is there any missing value remaining
df.isnull().sum()
```

```python
df_cleaned = df.copy()
```

```python
#Create a list of side effect columns
side_effect_columns = [col for col in df_cleaned.columns if 'sideEffect' in col]
df_cleaned[side_effect_columns].head(3)
```

```python
# Calculate the count of known side effects for each drug by counting entries that are not "Unknown"
df_cleaned['total_side_effect_recorded'] = df_cleaned[side_effect_columns].apply(lambda x: (x != "UnKnown").sum(), axis=1)
df_cleaned.head(3)
```

```python
#Top 10 Medicine that has most Side effects
df_cleaned[['name','total_side_effect_recorded']].sort_values(by ='total_side_effect_recorded',ascending=False).head(10)
```

```python
#Create a list of substitute columns
substitute_columns = [col for col in df_cleaned.columns if 'substitute' in col]
df_cleaned[substitute_columns].head(3)
```

```python
df_cleaned["total_substitute_counts"]=df_cleaned[substitute_columns].apply(lambda x: (x!= 'UnKnown').sum(),axis= 1)
df_cleaned.head(3)
```

```python
# Count non-"Unknown" values for each substitute column
substitute_counts = pd.DataFrame(df_cleaned[substitute_columns].apply(lambda col: (col != "UnKnown")).sum(axis = 0),columns= ['Counts']).reset_index()
substitute_counts=substitute_counts.rename(columns={'index':'substitutes'})
substitute_counts
```

```python
#creating a line with the missing value

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 4))
sns.lineplot(x='substitutes', y='Counts', data=substitute_counts, color ='r')

plt.xlabel('Substitutes')
plt.ylabel('Substitutes Count')
plt.title("Substitutes VS Substitutes Counted")
plt.xticks(rotation=0)


# Show the plot
plt.show()
```

```python
df_cleaned['total_uses'] = df_cleaned[use_columns].apply(lambda x: x !='UnKnown').sum(axis =1)
df_cleaned.sort_values(by= 'total_uses',ascending = False).head(3)
```

```python
# Combine all use columns into one
use_columns = [f'use{i}' for i in range(5)]

uses = df_cleaned[use_columns].apply(lambda x: x.value_counts()).sum(axis =1)
uses = pd.DataFrame(uses,columns = ['uses_counts']).reset_index()
uses = uses.rename(columns = {'index':'uses'})

# Filter out rows where 'uses' is "Unknown" or null
uses = uses[uses.apply(lambda x: (x!= 'UnKnown'))]
uses = uses.dropna()

#sorting the dataframe
uses = uses.sort_values("uses_counts",ascending = False)

#Top 10 using
uses.head(10)
```
